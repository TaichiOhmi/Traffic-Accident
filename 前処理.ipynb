```python
import re

def multiple_replace(text, adict):
    rx = re.compile('|'.join(adict))
    def dedictkey(text):
        for key in adict.keys():
            if re.search(key, text):
                return key
 
    def one_xlat(match):
        return adict[dedictkey(match.group(0))]
 
    return rx.sub(one_xlat, text)
```

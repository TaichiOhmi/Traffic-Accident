```python
def tasuzero(month, day):
    if month==1:
        month='01'
    elif month ==2:
        month='02'
    elif month==3:
        month='03'
    elif month==4:
        month='04'
    elif month==5:
        month='05'
    elif month==6:
        month='06'
    elif month==7:
        month='07'
    elif month==8:
        month='08'
    elif month==9:
        month='09'
    else:
        month = month
    if day==1:
        day='01'
    elif day ==2:
        day='02'
    elif day==3:
        day='03'
    elif day==4:
        day='04'
    elif day==5:
        day='05'
    elif day==6:
        day='06'
    elif day==7:
        day='07'
    elif day==8:
        day='08'
    elif day==9:
        day='09'
    else:
        day = day
    return month, day
```
